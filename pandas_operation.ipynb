## merging 2 dataframes

```python
#merging 2 dataframes
import numpy as np
import pandas as pd
```

```python
# dataframe 1 employee info
employee= {
    'employee_id': [1,2,3,4,5],
    'Name':['john','anna','peter','linda','bob'],
    'department':['hr','it','finance','it','hr']
}
emp = pd.DataFrame(employee)

#dataframe 2 salary info
salaries={
    'employee_id': [1,2,3,6,7],
    'salary':[60000,80000,65000,70000,90000],
    'bonus':[5000,10000,7000,8000,12000]
}

sal= pd.DataFrame(salaries)
```

```python
emp
```

```python
sal
```

```python
pd.merge(emp,sal,on= 'employee_id', how='inner')
#gives common values
```

```python
pd.merge(emp,sal,on='employee_id',how='outer')
#includes all ids of both table
```

```python
pd.merge(emp,sal,on='employee_id',how='left')
#include all ids of left or 1st table
```

```python
pd.merge(emp,sal,on='employee_id',how='right')
#include all ids of right or 2nd table
```

## Group by 

```python
data1= {
    'category':['A','B','A','B','A','B','A','B'],
    'store':['s1','s1','s2','s2','s1','s2','s2','s1'],
     'sales':[100,200,150,250,120,180,200,300],
     'quantity':[10,15,12,18,8,20,15,25],
'date':pd.date_range('2023-01-01', periods=8)
}
grp= pd.DataFrame(data1)
```

```python
grp
```

```python
#group by category and calculate the sum of sales
cat = grp.groupby('category')
cat
for i,v in cat:
    print(i)
    print(v) 
#this is only for how groups are created , optional
```

```python
total=grp.groupby('category')['sales'].sum()
total
```

```python
#group by store and calculate the sum of sales
grp.groupby('store')['sales'].sum()
```

```python
#group by multiple columns then write in form of list 
grp.groupby(['category','store'])['sales'].sum()
```

## pivot tables

```python
data2 = {
    'Date': pd.date_range('2023-01-01', periods=20),
    'Product': ['A', 'B', 'C', 'D'] * 5,
    'Region': ['East', 'West', 'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West',
               'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North', 'South'],
    'Sales': np.random.randint(100, 1000, 20),
    'Units': np.random.randint(10, 100, 20),
    'Rep': ['John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary',
            'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice']
}
piv = pd.DataFrame(data2)
```

```python
piv
```

```python
piv['Month'] = piv['Date'].dt.month_name()
piv['Quarter']= 'Q'+ piv['Date'].dt.quarter.astype(str)
piv
```

```python
pd.pivot_table(piv,values= 'Sales', index='Region', columns='Product',aggfunc='mean')
```

```python
#shift + tab to know the attributes of pivot_table
#by default table shows mean of sales , we can change it to median or other
```

```python
pd.pivot_table(piv,values=['Sales','Units'],index='Product',columns='Region',aggfunc='median')
#the values attribute  should have the numeric values
```

```python
#cross Tab
# used for the counting purpose 

pd.crosstab(piv['Region'],piv['Product'])
```
